```python
Calls["Reason"]= Calls["title"].apply(lambda title: title.split(":")[0])
```

```python
Calls.head()
```

```python
Calls["Reason"].value_counts()
```

```python
sns.countplot(x="Reason", data=Calls, palette="viridis")
```

```python
Calls.info()
```

```python
type(Calls["timeStamp"].iloc[0])
```

```python
Calls["timeStamp"]= pd.to_datetime(Calls["timeStamp"])
```

```python
type(Calls["timeStamp"].iloc[0])
```

```python
Calls["Hour"] =Calls["timeStamp"].apply(lambda time : time.hour) 
Calls["Month"]= Calls["timeStamp"].apply(lambda time : time.month)
Calls["Day of Week"] = Calls["timeStamp"].apply(lambda time : time.dayofweek)
```

```python
Calls.head()
```

```python
Calls["Day of Week"]
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
Calls["Day of Week"]= Calls["Day of Week"].map(dmap)
```

```python
Calls.head()
```

```python
sns.countplot(x="Day of Week", data= Calls, hue="Reason", palette="viridis")
```

```python
sns.countplot(x="Month", data=Calls, hue="Reason", palette="viridis")
```

```python
byMonth = Calls.groupby('Month').count()
byMonth
```

```python
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
Calls['Date']=Calls['timeStamp'].apply(lambda t: t.date())
```

```python
Calls.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
Calls['Month'].unique()
```

```python
Calls[Calls['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
Calls[Calls['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
Calls[Calls['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
dayHour = Calls.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

```python
dayMonth = Calls.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
sns.clustermap(dayMonth,cmap='viridis')
```
